```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
students = ['Ali', 'Sara', 'Mona', 'Omar']
subjects = ['Math', 'English', 'Physics']
grades = np.array([
    [85, 78, 92],  
    [90, 88, 75],  
    [60, 70, 65],  
    [95, 85, 80]   
])
```

```python
print(" Average, Highest, and Lowest scores per student:")

for i, student in enumerate(students):
    avg_score = np.mean(grades[i])
    max_score = np.max(grades[i])
    min_score = np.min(grades[i])
    print(f"{student}: Avg = {avg_score:.2f}, Highest = {max_score}, Lowest = {min_score}")
```

```python
pass_count = 0
fail_count = 0
for i in range(len(students)):
    if np.mean(grades[i]) >= 70:
        pass_count += 1
    else:
        fail_count += 1

print(f"\n Number of passing students: {pass_count}")
print(f" Number of failing students: {fail_count}")
```

```python
for i, student in enumerate(students):
    plt.figure(figsize=(6,4))
    plt.bar(subjects, grades[i], color='skyblue')
    plt.ylim(0, 100)
    plt.title(f"{student}'s Grades")
    plt.xlabel("Subject")
    plt.ylabel("Score")
    for j, score in enumerate(grades[i]):
        plt.text(j, score + 1, str(int(score)), ha='center')
    plt.show()
```
